```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
Indices = ['NAS@IXIC', 'SHS@000001', 'LNS@FTSE100', 'PAS@CAC40', 'XTR@DAX30', 'STX@SX5E']
```

```python
nasdaq_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
shanghai_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
london_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
paris_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
dutch_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
euro_df = pd.DataFrame(columns=['date', 'close', 'start', 'high', 'low'])
```

```python
for index in Indices : 
    for k in range(1,155) :
        url = 'https://finance.naver.com/world/sise.naver?symbol={}&page={}'.format(index,k)
        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')
        path = '#content > div.section_quot'
        index_table = soup.select(path)

        for i in range(1,11) :
            date = index_table[0].select('.tb_td')[i].text
            close = index_table[0].select('.tb_td2')[i].text
            start = index_table[0].select('.tb_td4')[i].text
            high = index_table[0].select('.tb_td5')[i].text
            low = index_table[0].select('.tb_td6')[i].text

            data = {'date' : date, 'close' : close, 'start' : start, 'high' : high, 'low' : low}
            if index == 'NAS@IXIC' :
                nasdaq_df = nasdaq_df.append(data, ignore_index=True)
            elif index == 'SHS@000001' :
                shanghai_df = shanghai_df.append(data, ignore_index=True)
            elif index == 'LNS@FTSE100' :
                london_df = london_df.append(data, ignore_index=True)
            elif index == 'PAS@CAC40' :
                paris_df = paris_df.append(data, ignore_index=True)
            elif index == 'XTR@DAX30' :
                dutch_df = dutch_df.append(data, ignore_index=True)
            elif index == 'STX@SX5E' :
                euro_df = euro_df.append(data, ignore_index=True)
```

```python
nasdaq_df.to_csv('./crawling_data/nasdaq_df.csv')
shanghai_df.to_csv('./crawling_data/shanghai_df.csv')
london_df.to_csv('./crawling_data/london_df.csv')
paris_df.to_csv('./crawling_data/paris_df.csv')
dutch_df.to_csv('./crawling_data/dutch_df.csv')
euro_df.to_csv('./crawling_data/euro_df.csv')
```
