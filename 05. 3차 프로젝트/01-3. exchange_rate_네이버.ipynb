```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from datetime import datetime
import json

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import time
```

```python
url = 'https://finance.naver.com/marketindex/exchangeDetail.naver?marketindexCd=FX_USDKRW'
```

```python
driver = webdriver.Chrome()
driver.get(url)
```

```python
iframe = driver.find_element_by_xpath('/html/body/div[3]/div[2]/div[2]/iframe[2]')
driver.switch_to_frame(iframe)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
path='body > div > table > tbody > tr'
table_data=soup.select(path)
```

```python
# 달러
USD = pd.DataFrame(columns=['date','매매기준율'])
for page_num in range(2,146): 
    
    url='https://finance.naver.com/marketindex/exchangeDailyQuote.naver?marketindexCd=FX_USDKRW&page={}'.format(page_num)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    path='body > div > table > tbody > tr'
    table_data=soup.select(path)
    
    for i in table_data:
        date = i.select('td')[0].text
        close = i.select('td')[1].text
        print('날짜 : {} , 매매기준율 : {},'.format(date, close))
        data = {'date' : date, '매매기준율' : close}
        USD = USD.append(data, ignore_index=True)
    print()
USD.to_csv('usd_list_naver.csv',encoding='utf-8-sig')
```
