## p d q 
1. p : 'AR'이 몇번째 과거까지 바라보는지
2. d : 차분한 횟수
3. q : MA가 몇번째 과거까지를 바라보는지
4. d = 0 ARMA 모형, 이 모형은 정상성을 만족한다.
5. p = 0 IMA 모형 d번 차분하면 MA 모형을 따르게 된다.
6. q = 0 ARI 모형 d번 차분하면 AR 모형을 따르게 된다.

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
model = ARIMA(df.iloc[:,-1], order = (3,2,3))    # p d q 
model_fit = model.fit(trend='c', full_output = True, disp = True)
model_fit.summary()
```

```python
fig = model_fit.plot_predict()
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
```
