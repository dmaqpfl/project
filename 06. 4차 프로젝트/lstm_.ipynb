```python
sc = StandardScaler()

df[['y_탄소배출권_선물_내역종가']] = sc.fit_transform(df[['y_탄소배출권_선물_내역종가']])
```

```python
price = df['y_탄소배출권_선물_내역종가'].values.tolist()
```

```python
window_size =5
X=[]
Y=[]

for i in range(len(price)- window_size):

    X.append([price[i+j]for j in range(window_size)])
    Y.append(price[window_size+i])
```

```python
x = np.asarray(X)
y = np.asarray(Y)
```

```python
print(x.shape)
print(y.shape)
```

```python
train_test_spilt =1000

X_train =x[:train_test_spilt,:]
Y_train =y[:train_test_spilt]
X_test =x[train_test_spilt:,:]
Y_test = y[train_test_spilt:]
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
X_train = np.reshape(X_train , (X_train.shape[0], window_size, 1))
X_test = np.reshape(X_test , (X_test.shape[0], window_size, 1))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = Sequential()
model.add(LSTM(15, input_shape=(28,1,)))
model.add(Dropout(0.2))
# model.add(Dense(16))
# model.add(Dense(5))
# model.add(Dense(3))
model.add(Dense(1, activation='linear'))  # 연속값 데이터 - 회귀분석
model.summary()

model.compile(loss='mse', optimizer='adam', metrics=['accuracy', 'mse'])
model.fit(X_train, Y_train,batch_size=1,epochs=30,validation_data=(X_test,Y_test))
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

plt.figure(figsize=(15,10))
plt.plot(price, color='orange')

split_pt= train_test_spilt + window_size
# train
plt.plot(np.arange(window_size, split_pt,1), train_predict, color='g')

# test
plt.plot(np.arange(split_pt, split_pt+len(test_predict),1), test_predict, color='r')
```
